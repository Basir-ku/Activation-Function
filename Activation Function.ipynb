# Relu

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate a range of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
# Apply the ReLU function to these values
y_values = relu(x_values)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='ReLU Function')
plt.title('ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

# LRelu

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Leaky ReLU function
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, x * alpha)

# Generate a range of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
# Apply the Leaky ReLU function to these values
y_values_lrelu = leaky_relu(x_values)

# Plot the Leaky ReLU function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_lrelu, label='Leaky ReLU Function', color='red')
plt.title('Leaky ReLU Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```

# Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate a range of values from -10 to 10
x_values = np.linspace(-10, 10, 400)
# Apply the Sigmoid function to these values
y_values_sigmoid = sigmoid(x_values)

# Plot the Sigmoid function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_sigmoid, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
